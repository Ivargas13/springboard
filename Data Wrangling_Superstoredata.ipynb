```python
#based on https://en.wikipedia.org/wiki/Southeastern_United_States the southeast region consitst of the listed states below:
southeast_region=  ['Alabama', 'Arkansas', 'Florida', 'Georgia', 'Kentucky', 
    'Louisiana', 'Maryland', 'Mississippi', 'North Carolina', 'South Carolina', 
    'Tennessee', 'Virginia', 'West Virginia']
```

```python
df_southeast = df[df['State'].isin(southeast_region)]
df_southeast.head()
```

```python
#seaborn boxplot of the df_southeast,
#with 'state' on the x-axis, 'Salse' as the y-value, and a hue that indicates 'Category'
#This will use boxplot's x, y, hue, and data arguments.
plt.subplots(figsize=(12, 8))
sns.boxplot(x='State', y='Sales', hue='Category', data=df_southeast)
plt.xticks(rotation='vertical')
plt.ylabel('Sales ($)')
plt.xlabel('State');
```

```python
#Sort category by sales for the southeast region
Top_category = df_southeast.groupby(["Category"]).sum().sort_values("Sales", ascending=False).head(10) 
Top_category = Top_category[['Sales']].round(2) 
```

```python
#Reset the index to add the category name into dataframe 
Top_category.reset_index(inplace=True) 
Top_category.head()
```

```python
#bar plot specific size 
plt.figure(figsize=(12, 8))
#using the Top ctegory as x-axis and Sales as the y axis 
sns.barplot(x='Category', y='Sales', data=Top_category, palette='viridis')
#customizing the title and name for the x and y axis 
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Top Sales Categories in the Southeast Region')
plt.xticks(rotation=45) 
#show plot
plt.tight_layout()
plt.show()
```

```python
#cisco teleprece has the most sales by more than half of the next top product in Technology HP designjet
#Get the corelating heatmap for the columns category and sales in df_southeast and compare the df
#f' with columns 'category' and 'Sales'
df_matrix = df_southeast[['Category', 'Sales']]
#Create dummy variables for the 'Category' column
df_cat = pd.get_dummies(df_matrix)
# Calculating the correlation matrix
cor_mat = df_cat.corr()

# Create the heatmap
fig = plt.gcf()
fig.set_size_inches(8, 6)
sns.heatmap(data=cor_mat, square=True, annot=True, cbar=True)
plt.title("Correlation Heatmap: Category vs. Sales")
plt.show()
```
